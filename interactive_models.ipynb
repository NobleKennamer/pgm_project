```python
import gensim
import pyLDAvis
import pyLDAvis.gensim as gensimvis
```

```python
def hdp_to_lda(hdp):
    alpha, beta = hdp.hdp_to_lda()
    hdp_lda = gensim.models.LdaModel(id2word=hdp.id2word,num_topics=len(alpha), alpha=alpha, eta=hdp.m_eta) 
    hdp_lda.expElogbeta = numpy.array(beta, dtype=numpy.float32) 
    return hdp_lda
```

```python
corpus = gensim.corpora.MmCorpus('models/corpus.mm')
dictionary = gensim.corpora.Dictionary.load('models/dictionary.dict')
lda = gensim.models.LdaModel.load('models/emails_lda.model')
hdp = gensim.models.HdpModel.load('models/emails_hdp.model')
hdp_lda = hdp_to_lda(hdp)
```

# LDA Visualization

```python
vis_data = gensimvis.prepare(lda, corpus, dictionary, R=10)
pyLDAvis.display(vis_data)
```

# HDP Visualization

```python
vis_data = gensimvis.prepare(hdp_lda, corpus, dictionary, R=10)
pyLDAvis.display(vis_data)
```
